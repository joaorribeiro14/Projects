```python
print("=======================================")
print("Bem Vindo ao jogo!")
print("=======================================")
import random
import os

def clear_screen():
    os.system('cls')


def jogo():
    computador = [(0,"Papel"),
                (1,"Pedra"),
                (2,"Tesoura")]
    pessoa = [(0,"Papel"),
            (1,"Pedra"),
            (2,"Tesoura")]
    placar_voce = 0
    placar_computador = 0
    while True:
        clear_screen()
        print("Você: {}".format(placar_voce))
        print("Computador: {}".format(placar_computador))
        print("Escolha seu lance:")
        for i,pesso in enumerate(pessoa):
            print("{} - {}".format(pesso[0],pesso[1]))
        escolha = int(input())
        pc = random.choice(computador)[0]
        print("Sua jogada: {}".format(pessoa[escolha][1]))
        print("Jogada do computador: {}".format(computador[pc][1]))
        
        
        if escolha == pc:
            print("O jogo deu empate! Deseja jogar novamente?")
        if escolha == 0 and pc == 1:
            print("Você venceu! Deseja jogar novamente?")
            placar_voce +=1
        if escolha == 0 and pc == 2:
            print("Você perdeu! Deseja jogar novamente?")
            placar_computador +=1
        if escolha == 1 and pc == 2:
            print("Você venceu! Deseja jogar novamente?")
            placar_voce +=1
        if escolha == 1 and pc == 0:
            print("Você perdeu! Deseja jogar novamente?")
            placar_computador +=1
        if escolha == 2 and pc == 0:
            print("Você venceu! Deseja jogar novamente?")
            placar_voce +=1
        if escolha == 2 and pc == 1:
            print("Você perdeu! Deseja jogar novamente?")
            placar_computador +=1
        print("0 - Continuar | 1 - Sair")
        saida = int(input())
        clear_screen()
        
        if saida == 0:
            continue
          
        else:
            print("Placar final:\nVocê: {}\nComputador: {}".format(placar_voce,placar_computador))
            if placar_voce>placar_computador:
                print("Você venceu!!!")
            elif placar_voce == placar_computador:
                print("Você empatou!!!")
            else:
                print("Você perdeu!!!")
            print("Obrigado por jogar! Até a próxima.")
            break
            clear_screen()
    
            

jogo()
```
