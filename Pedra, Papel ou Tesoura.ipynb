```python
import os
import random

class TicTacToe:
    def __init__(self):
        self.board = [[" "," "," "],[" "," "," "],[" "," "," "]]
        self.done = ""
        
    def print_board(self):
        print("=============================")
        print("Bem-vindo(a) ao Jogo da Velha")
        print("=============================")
        print("INSTRUÇÕES")
        print("A cada jogada, iremos pedir que você indique um número entre 1 e 9.")
        print("Os números correspondem à posição em que você quer que a jogada seja feita.")
        print("Veja abaixo a posição dos números no tabuleiro:")
        print("")
        print(" 1 | 2 | 3 ")
        print("-----------")
        print(" 4 | 5 | 6 ")
        print("-----------")
        print(" 7 | 8 | 9 ")
        print("A cada rodada, você poderá inserir a sua opção apenas nos espaços vazios.")

    def reset(self):
        self.board = [[" "," "," "],[" "," "," "],[" "," "," "]]
        self.done = ""

    def verifica_vitoria(self):
        # Combinações vitoriosas
        for i in ["X","O"]:
            if ((self.board[0][0] == self.board[0][1] == self.board[0][2] == i) or
                (self.board[1][0] == self.board[1][1] == self.board[1][2] == i) or
                (self.board[2][0] == self.board[2][1] == self.board[2][2] == i) or
                (self.board[0][0] == self.board[1][0] == self.board[2][0] == i) or
                (self.board[0][1] == self.board[1][1] == self.board[2][1] == i) or
                (self.board[0][2] == self.board[1][2] == self.board[2][2] == i) or
                (self.board[0][0] == self.board[1][1] == self.board[2][2] == i) or
                (self.board[0][2] == self.board[1][1] == self.board[2][0] == i)):
                self.done = i
                print(f"{i} Venceu")
                return

        # Verifica empate
        if all(cell != " " for row in self.board for cell in row):
            self.done = 'd'
            print("Empate!")
            return
    
    def player_move(self):
        while True:
            try:
                move = int(input("Digite um número entre 1 e 9: "))
                if move < 1 or move > 9:
                    print("Coordenadas inválidas. Tente novamente.")
                    continue
                
                x, y = divmod(move - 1, 3)
                if self.board[x][y] != " ":
                    print("Posição já preenchida. Tente novamente.")
                else:
                    self.board[x][y] = "X"
                    break
            except ValueError:
                print("Entrada inválida. Digite um número entre 1 e 9.")

    def pc_move(self):
        pc_moves = [(i, j) for i in range(3) for j in range(3) if self.board[i][j] == " "]
        
        if pc_moves:
            x, y = random.choice(pc_moves)
            self.board[x][y] = "O"
    
    def display_board(self):
        
        print("")
        print(" " + self.board[0][0] + " | " + self.board[0][1] + " | " + self.board[0][2])
        print("-----------")
        print(" " + self.board[1][0] + " | " + self.board[1][1] + " | " + self.board[1][2])
        print("-----------")
        print(" " + self.board[2][0] + " | " + self.board[2][1] + " | " + self.board[2][2])

# Executa o jogo
tic_tac_toe = TicTacToe()
next_game = 0

while next_game == 0:
    os.system('cls' if os.name == 'nt' else 'clear')
    tic_tac_toe.print_board()
    tic_tac_toe.display_board()
    while tic_tac_toe.done == "":
        
        tic_tac_toe.player_move()
        tic_tac_toe.pc_move()
        tic_tac_toe.display_board()
        tic_tac_toe.verifica_vitoria()
    
    print("Digite 1 para sair ou qualquer outra tecla para jogar novamente")
    next_game = int(input())
    if next_game != 1:
        tic_tac_toe.reset()
        os.system('cls' if os.name == 'nt' else 'clear')
        tic_tac_toe.print_board()
        next_game = 0
    else:
        os.system('cls' if os.name == 'nt' else 'clear')
```
