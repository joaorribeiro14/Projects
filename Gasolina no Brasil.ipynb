### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df_gas['ANO'] = df_gas["DATA FINAL"].apply(lambda x: x.year)
```

```python
df_rio = df_gas[['ESTADO','ANO','PREÇO MÉDIO REVENDA']][df_gas['ESTADO'] == 'RIO DE JANEIRO'].groupby('ANO')['PREÇO MÉDIO REVENDA'].sum()
```

```python
df_rio
```

```python
(df_rio/df_rio.shift(1) - 1) *100
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
caro = df_gas.groupby('ANO')[['PREÇO MÉDIO REVENDA','ESTADO']].max().reset_index()
```

```python
barato = df_gas.groupby('ANO')[['PREÇO MÉDIO REVENDA','ESTADO']].min().reset_index()
```

```python
misto = pd.concat([caro,barato]).sort_index()
```

```python
misto
```

```python
misto['VARIAÇÃO%'] = (misto['PREÇO MÉDIO REVENDA']/misto['PREÇO MÉDIO REVENDA'].shift(1) - 1) *100
```

```python
misto['DIF ABSOLUTA'] = (misto['PREÇO MÉDIO REVENDA']- misto['PREÇO MÉDIO REVENDA'].shift(1))
```
